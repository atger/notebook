```python
line1_func = lambda p: (((1-p)**2)-0.03, (p**2)-0.03)
```

```python
line2_func = lambda p: (((1-p)**2)-0.03, (p**2)+0.03)
```

```python
line1 = [i for i in map(line1_func,p)]
```

```python
line2 = [i for i in map(line2_func,p)]
```

```python
is_above = lambda p1, p2: p1[0] > p2[0] and p1[1] > p2[1]
```

```python
is_below = lambda p1, p2: p1[0] < p2[0] and p1[1] < p2[1]
```

```python
sum([is_below((0.1,0),p) for p in line1 ])
```

```python
point_below_line = lambda p, line: True if sum([is_below(p,point) for point in line ]) else False
```

```python
points = df[['level1', 'level2']].values.tolist()
```

```python
points[:10]
```

```python
[i for i in filter(lambda p: point_below_line(p,line1),points)]
```
